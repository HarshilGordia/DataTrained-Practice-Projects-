```python
df=df[(zscr<3).all(axis=1)]
```

```python
df.skew()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
df=pt.fit_transform(df)
```

```python
df=pd.DataFrame(df)
```

```python
df.columns=['AveragePrice', '4046', '4225', '4770', 'Small_Bags', 'Large_Bags',
       'XLarge_Bags', 'type']
```

```python
df.skew()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df)
df_avocado_scaler = scaler.transform(df)
df = pd.DataFrame(df_avocado_scaler)
df.columns = ['AveragePrice', '4046', '4225', '4770', 'SmallBags', 'LargeBags', 'XLargeBags', 'Type']
df.head()
```

```python
x=df.drop(columns='AveragePrice')
y=df['AveragePrice']
```

```python
x
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg=LinearRegression()
lg.fit(x_train,y_train)
```

```python
lg.coef_
```

```python
lg.intercept_
```

```python
lg.score(x_train,y_train)
```

```python
pred=lg.predict(x_test)
```

```python
print('R2 Score:',r2_score(y_test,pred))
print('\n')

print('error')
print('Mean absolute error:', mean_absolute_error(y_test,pred))
print('Mean squared error:', mean_squared_error(y_test,pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lg,x,y,cv=10,scoring='neg_mean_squared_error')
lg_score_cross=np.sqrt(-score)
np.mean(lg_score_cross),np.std(lg_score_cross)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
predrf=rf.predict(x_test)
```

```python
print('R2 Score:',r2_score(y_test,predrf))
print('\n')

print('error')
print('Mean absolute error:', mean_absolute_error(y_test,predrf))
print('Mean squared error:', mean_squared_error(y_test,predrf))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,predrf)))
```

```python
score=cross_val_score(rf,x,y,cv=10,scoring='neg_mean_squared_error')
rf_score_cross=np.sqrt(-score)
np.mean(rf_score_cross),np.std(rf_score_cross)
```
