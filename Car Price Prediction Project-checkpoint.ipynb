```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Used_Cars_Dataset')
df.head()
```

```python
df.shape
```

```python
df=df.drop(['Unnamed: 0'], axis='columns')
df
```

```python
df['Owner'].unique()
```

```python
df['Owner'].replace(to_replace = 'Owner\n1st', value =  '1',inplace=True)
df['Owner'].replace(to_replace = 'Owner\n2nd', value =  '2',inplace=True)
df['Owner'].replace(to_replace = 'Owner\n3rd', value =  '3',inplace=True)
df['Owner'].replace(to_replace = 'Owner\n4th', value =  '4',inplace=True)
df['Owner'].replace(to_replace = 'Owner\nFirst', value =  '1',inplace=True)
df['Owner'].replace(to_replace = 'Owner\nSecond', value =  '2',inplace=True)
df['Owner'].replace(to_replace = 'Owner\nThird', value =  '3',inplace=True)
df['Owner'].replace(to_replace = 'Owner\nFourth', value =  '4',inplace=True)
df['Owner'].replace(to_replace = 'Owner\n4+', value =  '4+',inplace=True)
df['Owner'].replace(to_replace = 'Owner\n--', value =  np.nan,inplace=True)
df['Owner'].replace(to_replace = '-', value = np.nan,inplace=True)
```

```python
df['Owner'].unique()
```

```python
df['Location'].value_counts()
```

```python
new = df["Location"].str.split(", ", n = 1, expand = True)
df["City"]= new[1]
df.drop(columns =["Location"], inplace = True)
df
```

```python
new = df["Model"].str.split(" ", n = 1, expand = True)
df["Brand"]= new[0]
df["Car Name"]= new[1]
df.drop(columns =["Model"], inplace = True)
df
```

```python
new = df["Kilometer"].str.split(" ", n = 1, expand = True)
df["Kilometers"]= new[0]
df.drop(columns =["Kilometer"], inplace = True)
df
```

```python
df.drop(columns =["Variant"], inplace = True)
df
```

```python
df.replace(to_replace = '-', value =  np.nan,inplace=True)
df.replace(to_replace = 'None', value =  np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
new = df["Price"].str.split(" ", n = 1, expand = True)
df["Car Price"]= new[1]
df.drop(columns =["Price"], inplace = True)
df
```

```python
new = df["Car Price"].str.split(",", n = 1, expand = True)
df["Car Price2"]= new[0]
df["Car Price1"]= new[1]
df.drop(columns =["Car Price"], inplace = True)
df
```

```python
new = df["Car Price1"].str.split(",", n = 1, expand = True)
df["Car Price3"]= new[0]
df["Car Price4"]= new[1]
df.drop(columns =["Car Price1"], inplace = True)
df
```

```python
Price = df['Car Price2'] + df['Car Price3'] + df['Car Price4']
Price
```

```python
df['Price']=Price
```

```python
df= df.drop(['Car Price2', 'Car Price3', 'Car Price4'], axis = 1)
df
```
