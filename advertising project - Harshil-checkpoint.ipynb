```python
x=df.drop(columns='sales')
y=df['sales']
```

```python
print(x)
print(x.shape)
```

```python
print(y)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error 
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
predlr=lr.predict(x_test)
```

```python
print(np.sqrt(mean_squared_error(y_test,predlr)))
```

```python
lr.score(x_train,y_train)
```

```python
print('predicted',predlr)
print('\n')
print('actual \n',y_test)
```

```python
print('error')

print('Mean absolute error:', mean_absolute_error(y_test,predlr))
print('Mean squared error:', mean_squared_error(y_test,predlr))

print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test,predlr)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,predlr))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(lr,x,y,cv=3)
print(score)
print(score.mean())
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.svm import SVC
svc=SVC(probability=True,kernel='linear')

#create adaboostclassifier object
ad=AdaBoostRegressor()
ad.fit(x_train,y_train)

ad_pred=ad.predict(x_test)
print(mean_squared_error(y_test,ad_pred))
print('Mean absolute error:', mean_absolute_error(y_test,ad_pred))
print(r2_score(y_test,ad_pred))
```

```python
adscores=cross_val_score(ad,x,y,cv=5)
print(adscores)
print(adscores.mean(),adscores.std())
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(mean_squared_error(y_test,predrf))
print('Mean absolute error:', mean_absolute_error(y_test,predrf))
print(r2_score(y_test,predrf))
```

```python
rfscores=cross_val_score(rf,x,y,cv=5)
print(rfscores)
print(rfscores.mean(),rfscores.std())
```

# Hyperparmeter Tuning

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'max_features': ['auto', 'log2']}
rf = RandomForestRegressor()
gr_search = GridSearchCV(estimator =rf,param_grid = param_grid,cv=5)
gr_search.fit(x_train, y_train)


print(gr_search.best_params_)
print(gr_search.best_score_)
```

```python
param_grid = {'n_estimators': [50, 100], 
                   'learning_rate': [0.01, 0.05, 0.1, 0.5], 
                   'loss': ['linear', 'square', 'exponential']},
ad= AdaBoostRegressor()
ad_gr_search = GridSearchCV(estimator =ad,param_grid = param_grid,cv=5)
ad_gr_search.fit(x_train, y_train)


print(ad_gr_search.best_params_)
print(ad_gr_search.best_score_)
```
